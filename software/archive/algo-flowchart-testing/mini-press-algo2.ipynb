```python
from diagrams import Diagram, Edge
from diagrams.programming.flowchart import StartEnd, Preparation, LoopLimit, InputOutput, Decision, Action
from IPython.display import Image

graph_attr = {
    'rankdir': 'TB',
    'fontsize': '12',
    'labeljust': 'r',
    'labelloc': 't',
}

edge_attr = {
    'len' : '0.0'
}

with Diagram("Microcontroller Control Loop", show=False, direction="TB", graph_attr=graph_attr, edge_attr = edge_attr):
    start = StartEnd("Start")
    setup = Preparation("Setup")
    loop_start = LoopLimit("Loop Start")
    read_input = InputOutput("Read Input")
    decision = Decision("Decision")
    action1 = Action("Action 1")
    action2 = Action("Action 2")
    continue_decision = Decision("Continue Loop?")
    loop_end = LoopLimit("Loop End")
    error_handling = Action("Error Handling")

    start >> setup >> loop_start >> read_input >> decision
    decision - Edge(label="Condition A", color="green") >> action1
    decision - Edge(label="Condition B", color="red") >> action2
    action1 >> continue_decision
    action2 >> continue_decision
    continue_decision - Edge(label="Yes", style="dashed") >> loop_start
    continue_decision - Edge(label="No") >> loop_end
    decision - Edge(label="Error", style="dotted", color="orange") >> error_handling
    error_handling >> loop_end

# Display the diagram
Image("Microcontroller_Control_Loop.png")
```

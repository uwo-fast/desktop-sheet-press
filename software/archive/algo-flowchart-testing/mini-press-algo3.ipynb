```python
from graphviz import Source
from IPython.display import Image

# DOT source code with added formatting
src = """
digraph heat_processing_control {
    rankdir=TB;

    // General styles
    node [fontname="Helvetica,Arial,sans-serif"];
    edge [fontname="Helvetica,Arial,sans-serif"];

    // Nodes with specific styles
    start [shape=ellipse, style=filled, fillcolor="#4CAF50", fontcolor=white, label="Start"];
    read_temp [shape=box, style=filled, fillcolor="#FFEB3B", label="Read Temperature"];
    compare_temp [shape=diamond, style=filled, fillcolor="#2196F3", fontcolor=white, label="Compare with Set Point"];
    heat_on [shape=box, style=filled, fillcolor="#F44336", fontcolor=white, label="Turn Heater ON"];
    heat_off [shape=box, style=filled, fillcolor="#8BC34A", label="Turn Heater OFF"];
    delay [shape=box, style=filled, fillcolor="#FFC107", label="Delay"];
    end [shape=ellipse, style=filled, fillcolor="#9E9E9E", fontcolor=white, label="End"];

    // Edges
    start -> read_temp;
    read_temp -> compare_temp;
    compare_temp -> heat_on [label="Below Set Point"];
    compare_temp -> heat_off [label="Above Set Point"];
    heat_on -> delay;
    heat_off -> delay;
    delay -> read_temp [label="Repeat"];
    read_temp -> end [label="Process Complete", style=dotted];
}
"""

s = Source(src, format="png")
# Render and display the image
img_path = s.render()  # This will save the image to a file
Image(filename=img_path)
```
